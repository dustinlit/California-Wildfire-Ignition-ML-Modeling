```python
import pandas as pd
```

```python
income = pd.read_csv('../data/raw/income.csv')
```

```python
cols = ['first','second','third','fourth']
```

```python
for col in cols:
    income[col] = income[col].str.replace('-', '0')
    income[col] = income[col].str.replace('+', '')
    income[col] = income[col].str.replace(',', '')
    income[col] = income[col].astype(int)
```

```python
income['Median Income'] = income['first'] + income['second'] + income['third'] + income['fourth']
```

```python
income[['tract','County','State']] = income['Geographic Area Name'].str.split(';', expand = True)
```
