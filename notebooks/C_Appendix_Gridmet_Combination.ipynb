```python
import pandas as pd
```

### Load all gridMET datasets

```python
weather = pd.read_csv('../data/raw/gridmet_point_extraction.csv')
spi180d_data = pd.read_csv('../data/raw/spi180d_extraction.csv')
spi30d_data = pd.read_csv('../data/raw/spi30d_extraction.csv')
spei30d_data = pd.read_csv('../data/raw/spei30d_extraction.csv')
spei90d_data = pd.read_csv('../data/raw/spei90d_extraction.csv')
spei180d_data = pd.read_csv('../data/raw/spei180d_extraction.csv')
pdsi_data = pd.read_csv('../data/raw/pdsi_extraction.csv')
names = pd.read_csv('../data/raw/gridmet_variables.csv')
```

```python
weather
```

## 1. Combine datasets

#### Rename index columns so they are unique

```python
spi180d_data.rename(columns={'spi': 'spi180d'}, inplace=True)
spi30d_data.rename(columns={'spi': 'spi30d'}, inplace=True)
spei30d_data.rename(columns={'spei': 'spei30d'}, inplace=True)
spei90d_data.rename(columns={'spei': 'spei90d'}, inplace=True)
spei180d_data.rename(columns={'spei': 'spei180d'}, inplace=True)
```

#### Attach data to main datset

```python
weather['spi30d'] = spi30d_data['spi30d']
weather['spi180d'] = spi180d_data['spi180d']
weather['spei30d'] = spei30d_data['spei30d']
weather['spei90d'] = spei90d_data['spei90d']
weather['spei180d'] = spei180d_data['spei180d']
weather['pdsi'] = pdsi_data['pdsi']
```

## 2. Clean Dataset and Meta File

```python
names
```

```python
# Create a units dictionary
units = {
    "tmmx": "°C",
    "tmmn": "°C",
    "vpd": "kPa",
    "rmax": "%",
    "rmin": "%",
    "sph": "kg/kg",
    "pr": "mm",
    "vs": "m/s",
    "srad": "W/m²",
    "etr": "mm",
    "bi": "index",
    "erc": "MJ/m²",
    "fm100": "%",
    "fm1000": "%",
    "spi30d": "index",
    "spi180d":"index",
    "spei30d":"index",
    "spei90d":"index",
    "spei180d":"index",
    "pdsi":"index"
}

# Map abbreviations to units
names["Units"] = names["Abbreviation"].map(units)
```

```python
New_Names = dict(zip(names['Abbreviation'], names['Full Name'] ))
```

```python
renamed = weather.rename(columns = New_Names)
```

```python
renamed
```

```python
weather_names = names[['Full Name','Units','Description']]
```

```python
weather_names
```

## 3. Export Files

```python
renamed.to_csv('../data/raw/gridmet_final.csv',index=False)
weather_names.to_csv('../data/raw/weather_meta.csv',index=False)
```
