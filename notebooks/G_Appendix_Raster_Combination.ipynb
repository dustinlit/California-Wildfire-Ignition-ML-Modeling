```python
import pandas as pd
```

```python
NDVI_2018 = pd.read_csv('../data/raw/NDVI_mean_2018.csv')
NDVI_2019 = pd.read_csv('../data/raw/NDVI_mean_2019.csv')
NDVI_2020 = pd.read_csv('../data/raw/NDVI_mean_2020.csv')
NDVI_2021 = pd.read_csv('../data/raw/NDVI_mean_2021.csv')
NDVI_2022 = pd.read_csv('../data/raw/NDVI_mean_2022.csv')
NDVI_2023 = pd.read_csv('../data/raw/NDVI_mean_2023.csv')
NDVI_2024 = pd.read_csv('../data/raw/NDVI_mean_2024.csv')
NDVI_2025 = pd.read_csv('../data/raw/NDVI_mean_2025.csv')
```

```python
NDVI_stats = pd.concat([NDVI_2018,
                       NDVI_2019,
                       NDVI_2020,                       
                       NDVI_2021,                       
                       NDVI_2022,                      
                       NDVI_2023,
                       NDVI_2024,                    
                       NDVI_2025                       
                       ])
```

```python
NDVI_stats['NDVI_mean_difference'] = NDVI_stats['NDVI_local_mean'] - NDVI_stats['NDVI_global_mean']
```

```python
NDVI_stats['NDVI_mean_difference'].describe()
```

```python
NDVI_stats['NDVI_mean_difference_has_value'] = NDVI_stats['NDVI_mean_difference'].notna().astype(int)
NDVI_stats['NDVI_mean_difference'] = NDVI_stats['NDVI_mean_difference'].fillna(0)
```

```python
NDVI_stats.isna().sum()
```

```python
NDVI_index = NDVI_stats[['Date','grid_id','NDVI_mean_difference','NDVI_mean_difference_has_value']]
```

```python
NDVI_index['NDVI_mean_difference_has_value'].value_counts()
```

```python
NDVI_index.to_csv('../data/raw/NDVI_index.csv', index = False)
print("All datasets saved successfully to ../data/processed/")
```
